```python
# Using data/ateneo.txt and data/tl-en.txt, convert the phrases into straight english phrases. 
#Leave words with no translation be.
```

```python
#import external files
with open("ateneo.txt") as f:
    ateneo_phrases = [i.strip('"') for i in f.read().split("\n")]
```

```python
ateneo_phrases
```

```python
with open("tl-en.txt", encoding = 'latin') as f:
    tl_en = dict([tuple(line.split("\t")) for line in f.read().split("\n") if line])
```

```python
tl_en
```

```python
#helpers
punctuations = [",", ".", "?"]

def capitalize(s):
    return s[0].toupper() + s[1:]

def isproper(s):
    return s.capitalize() == s

def haspunctuation(s):
    return s[-1] in punctuations
```

```python
def translate(sentence):
    words = sentence.split()
    output = []
    for word in words:
        temp_word = punc = ""
        if haspunctuation(word):
            temp_word = word[:-1].lower()
            punc = word[-1]
        else:
            temp_word = word.lower()
            
        translated_word = tl_en[temp_word] if temp_word in tl_en else temp_word
        
        output.append((translated_word.capitalize() if isproper(word) else translated_word) + punc)

    return " ".join(output)
```

```python
for i in ateneo_phrases:
    print(i)
    print(translate(i))
    print("-"*10)
```
