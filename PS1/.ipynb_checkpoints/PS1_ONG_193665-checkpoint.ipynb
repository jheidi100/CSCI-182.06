```python
def is_palindrome(word): 
    print(True if word[::-1].lower() == word.lower() else False)
```

```python
def is_prime(n):
    if n > 1:
        for i in range(2,n): 
            if n%i == 0:
                return(False)
        return(True)
    else: 
        return(False)
```

```python
def identify_keys(part):
    global v,c
    if "verse" in part.lower(): 
        key_name = "verse" + str(v)
        v+=1
    elif "chorus" in part.lower(): 
        key_name = "chorus" + str(c)
        c+=1
    return key_name

with open ("allstar.txt","r") as f:
    allstar_string = f.read()
    
allstar_list = allstar_string.split("[")
allstar_list.pop(0)
section_chords = dict()
v = 1
c = 1

for part in allstar_list: 
    key_name = identify_keys(part)
    part = part.replace("\n \n", "\n")
    part = part.split("\n")
    part.pop(0)
    part = part[::2]
    chord_list = []
    for line in part: 
        chord_list.extend(line.split())
    chord_string = ",".join(chord_list)

    section_chords[key_name] = chord_string
```

```python
def word_counter(word_count, long_string): 
    for word in long_string.split(): 
        if word not in word_count.keys(): 
            word_count[word]=1
        else: 
            word_count[word] += 1
    return word_count
            
```

```python
with open ("allstar.txt","r") as f:
    allstar_string = f.read()
    
allstar_string.replace("\n \n", "\n")
allstar_list = allstar_string.split("\n")
allstar_list = [line for line in allstar_list if not("[" in line or line == " ")]
allstar_list = allstar_list[1::2]
lyric_string = " ".join(allstar_list).lower()
word_count = dict()

word_counter(word_count, lyric_string)
```

```python
from bs4 import BeautifulSoup 
with open ("Rick Astley - Never Gonna Give You Up Lyrics _ AZLyrics.com.html") as f: 
    soup=BeautifulSoup(f)

parent_div = soup.find('div', {"class": 'col-xs-12 col-lg-8 text-center'})
lyrics_div = parent_div.find('div', {"class": None})
lyrics = lyrics_div.get_text()
```

```python
lyric_wordlist = lyrics.split()
lyric_linelist = lyrics.split("\n")

words = [word.strip("(), ").lower() for word in lyric_wordlist]
unique_words = len(set(words))
print(unique_words)

never_do_list = [line for line in lyric_linelist if line.startswith("Never gonna ")]
never_do = list(set(never_do_list))
print(never_do)

word_count2 = dict()
lyric_string = " ".join(words)
word_counter(word_count2, lyric_string)
```
